```python
import numpy as np
from torch import load
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
parameters = load("./params")
W = parameters[0]
W1 = W['W1']
W2 = W['W2']
print(W1.shape, W2.shape)
```

```python
pca = PCA(n_components=10)
newW1 = pca.fit_transform(W1)
print(newW1.shape)
```

Visualization of W1

```python
fig, axes = plt.subplots(2, 5)
scale = np.abs(newW1).max()
for i in range(2):
    for j in range(5):
        idx = j + i * 5
        axes[i, j].imshow(newW1[:, idx].reshape(28, 28),interpolation='nearest', cmap=plt.cm.Greys, vmin=-scale, vmax=scale)
        axes[i, j].axis('off')

plt.show()
```

Visualization of W2

```python
fig, axes = plt.subplots(2, 5)
newW2 = W2[:-3,:].copy()
scale = np.abs(newW2).max()
for i in range(2):
    for j in range(5):
        idx = j + i * 5
        axes[i, j].imshow(newW2[:, idx].reshape(16, 17),interpolation='nearest', cmap=plt.cm.Greys, vmin=-scale, vmax=scale)
        axes[i, j].axis('off')

plt.show()
```

Visualization of $W = W1\cdot W2$

```python
W = W1 @ W2
print(W.shape)
```

```python
fig, axes = plt.subplots(2, 5)
weights = W[:,:].copy()
scale = np.abs(weights).max()
for i in range(2):
    for j in range(5):
        idx = j + i * 5
        axes[i, j].imshow(weights[:, idx].reshape(28, 28),interpolation='nearest', cmap=plt.cm.Greys, vmin=-scale, vmax=scale)
        axes[i, j].axis('off')

plt.show()
```
